Pengujian Hipotesis (Anova Test)
Hipotesis :

H0 : Tidak cukup bukti menyatakan education tidak berpengaruh terhadap income.
H1 : Cukup bukti menyatakan education berpengaruh terhadap income.

```python
#import Library
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
#import Dataset
df = pd.read_csv('marketing_campaign.csv',sep="\t")
df.head()
```

```python
#mencari missing value
df.isna()
```

```python
df.notna()
```

```python
#output isna() dan notna() sulit dipahami karena terlalu banyak kolom dan baris
#maka lakukan agregasi data dengan fungsi sum()
#dengan agregasi data diketahui bahwa semua kolom dataset tidak terdapat nilai yang kosong kecuali pada kolom Income, terdapat 24 missing value.

df.isna().sum()
```

```python
#menangani missing value pada kolom Income dengan nilai rata-rata dari semua nilai income
df_income = df['Income'].mean() #mencari nilai rata-rata Income
df['Income'] = df['Income'].fillna(df_income) #mengisi nilai kosong dengan nilai rata-rata yang diperoleh
df['Income'].isna().sum()#mengecek kembali missing value di kolom Income
```

```python
df.isna().sum()
```

```python
#melihat jumlah value pada kolom Education
df['Education'].value_counts()
```

```python
#memunculkan Dataframe baris pertama
df.head()
```

```python
#memunculkan Dataframe baris terakhir
df.tail()
```

```python
#Memisahkan setiap kategori dalam Education
Graduation = df[df['Education']=='Graduation']
PhD = df[df['Education']=='PhD']
Master = df[df['Education']=='Master']
Second_Cycle = df[df['Education']=='2n Cycle']
Basic = df[df['Education']=='Basic']
```

```python
#melihat rata-rata Income untuk setiap kategori dalam Education
df.groupby(['Education'])['Income'].mean()
```

```python
#pengujian Anova
anova_test = st.f_oneway(Graduation['Income'], 
            PhD['Income'],
            Master['Income'],
            Second_Cycle['Income'],
            Basic['Income'])
```

```python
#Nilai Pvalue
anova_test.pvalue
```

```python
#Pengambilan keputusan
if anova_test.pvalue > 0.05:
    print('Tidak cukup bukti menyatakan education berpengaruh terhadap income')
else:
    print('Cukup bukti menyatakan education berpengaruh terhadap income')
        
```
